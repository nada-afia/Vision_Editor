```python
import cv2
import matplotlib.pyplot as plt
```

```python
def show_grid(image_path):

    img = cv2.imread(image_path)
    if img is None:
        print("Image not found")
        return

    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    equ = cv2.equalizeHist(gray)

    plt.figure(figsize=(12, 10))

    plt.subplot(4, 2, 1)
    plt.imshow(img_rgb)
    plt.title("Original Image")
    plt.axis('off')

    plt.subplot(4, 2, 2)
    for i, col in enumerate(('r','g','b')):
        hist = cv2.calcHist([img],[i],None,[256],[0,256])
        plt.plot(hist, color=col)
    plt.title("Original Histogram")

    plt.subplot(4, 2, 3)
    plt.imshow(gray, cmap='gray')
    plt.title("Gray Image")
    plt.axis('off')

    plt.subplot(4, 2, 4)
    hist_gray = cv2.calcHist([gray],[0],None,[256],[0,256])
    plt.plot(hist_gray, color='pink')
    plt.title("Gray Histogram")

    plt.subplot(4, 2, 5)
    plt.imshow(equ, cmap='gray')
    plt.title("Equalized Image")
    plt.axis('off')

    plt.subplot(4, 2, 6)
    hist_equ = cv2.calcHist([equ],[0],None,[256],[0,256])
    plt.plot(hist_equ, color='pink')
    plt.title("Equalized Histogram")

    img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
    img_yuv[:, :, 0] = cv2.equalizeHist(img_yuv[:, :, 0])
    img_output = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)

    plt.subplot(4, 2, 7)
    plt.imshow(cv2.cvtColor(img_output, cv2.COLOR_BGR2RGB))
    plt.title("Color Equalized")
    plt.axis('off')

    plt.subplot(4, 2, 8)
    for i, col in enumerate(('r','g','b')):
        hist = cv2.calcHist([img_output],[i],None,[256],[0,256])
        plt.plot(hist, color=col)
    plt.title("Color Equalized Histogram")

    plt.tight_layout()
    plt.show()


# RUN
show_grid(r"C:\Users\nadam\Desktop\project\image\cats.jpg")
```
